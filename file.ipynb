```python
import math
pre_cost = 2.0E+7
fix_cost = 1000*1.0E+4
cost = pre_cost + fix_cost

def calc_earnings(pre_cost):
    earnings = 2.87E+07 * math.log(pre_cost)-4.44E+08
    return int(earnings)
earnings = calc_earnings(pre_cost)

def calc_profit(earnings,cost):
    profit = earnings - cost
    return int(profit)
profit = calc_profit(earnings,cost)
profit_ratio = int(profit/cost*100)
data_ad_cost = list(range(1000,9000,1))
data_earnings = [calc_earnings(ad_cost*1.0E+04) for ad_cost in data_ad_cost]
data_profit = [calc_profit(earnings,ad_cost*1.0E+04+fix_cost) for earnings,ad_cost in zip(data_earnings,data_ad_cost)]

import matplotlib.pyplot as plt
plt.plot(data_ad_cost,data_earnings,color = 'blue')
plt.plot(data_ad_cost,data_profit,color = 'green')
plt.show()

max_profit = max(data_profit)
best_ad_cost = data_ad_cost[data_profit.index(max_profit)]
best_ad_cost
```
